```python
from ucimlrepo import fetch_ucirepo 
```

```python
statlog_heart = fetch_ucirepo(id=145)
```

```python
x = statlog_heart.data.features 
y = statlog_heart.data.targets 
```

```python
x
```

```python
y
```

```python
df = x
```

```python
df['target']=y
```

```python
df.head()
```

```python
df.isnull().sum()
```

```python
from sklearn.model_selection import train_test_split 
X_train ,X_test,y_train,y_test = train_test_split(df.drop(['target'],axis='columns'),df['target'],test_size=0.3,random_state=42)
```

```python
len(X_train)
```

```python
len(X_test)
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
lr = LogisticRegression()
```

```python
lr.fit(X_train,y_train)
```

```python
lr.score(X_test,y_test)
```

```python
y_pred_lr = lr.predict(X_test)
```

```python
from sklearn.metrics import classification_report
print(classification_report(y_test,y_pred_lr))
```

```python
from sklearn.ensemble import RandomForestClassifier
rf = RandomForestClassifier(n_estimators=200,
    max_depth=10,
    min_samples_split=5,
    random_state=42)
```

```python
rf.fit(X_train,y_train)
```

```python
rf.score(X_test,y_test)
```

```python
y_pred_rf = rf.predict(X_test)
```

```python
print(classification_report(y_test,y_pred_rf))
```

```python
from sklearn.svm import SVC
svc = SVC(C=13.0,gamma='scale')
```

```python
svc.fit(X_train,y_train)
```

```python
svc.score(X_test,y_test)
```

```python
y_pred_svc = svc.predict(X_test)
```

```python
print(classification_report(y_test,y_pred_svc))
```
